**Alzheimer's Disease**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define the values
true_positives = 444
true_negatives = 895
false_positives = 6
false_negatives = 6

# Create the confusion matrix
confusion_matrix = np.array([
    [true_positives, false_negatives],
    [false_positives, true_negatives]
])

# Plot the confusion matrix
plt.figure(figsize=(5, 4))
sns.heatmap(confusion_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Positive', 'Negative'], yticklabels=['Positive', 'Negative'])

# Add titles and labels
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

**Combined Confusion Matrix for the AD, CN and CI**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define the values for AD, CI, and CN
# Format: [True Positives, False Negatives]
# Format: [False Positives, True Negatives]

# Alzheimer Disease (AD)
AD_tp = 444
AD_tn = 895
AD_fp = 6
AD_fn = 6

# Cognitive Impairment (CI)
CI_tp = 444
CI_tn = 890
CI_fp = 11
CI_fn = 6

# Cognitively Normal (CN)
CN_tp = 436
CN_tn = 890
CN_fp = 10
CN_fn = 15

# Create the confusion matrix for AD, CI, and CN
confusion_matrix = np.array([
    [AD_tp, AD_fn, 0],  # AD - True Positives and False Negatives
    [CI_fp, CI_tp, CI_fn],  # CI - False Positives, True Positives, False Negatives
    [CN_fp, 0, CN_tp]  # CN - False Positives, True Negatives
])

# Plot the confusion matrix
plt.figure(figsize=(5, 4))
sns.heatmap(confusion_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['AD', 'CI', 'CN'], yticklabels=['AD', 'CI', 'CN'])

# Add titles and labels
plt.title('Confusion Matrix for AD, CI, and CN')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define the values for AD, CI, and CN
# Alzheimer Disease (AD)
AD_tp = 444
AD_tn = 895
AD_fp = 6
AD_fn = 6

# Cognitive Impairment (CI)
CI_tp = 444
CI_tn = 890
CI_fp = 11
CI_fn = 6

# Cognitively Normal (CN)
CN_tp = 436
CN_tn = 890
CN_fp = 10
CN_fn = 15

# Create the confusion matrix with all metrics
confusion_matrix = np.array([
    [f"TP: {AD_tp}", f"FP: {AD_fp}", f"FN: {AD_fn}", f"TN: {AD_tn}"],  # AD
    [f"TP: {CI_tp}", f"FP: {CI_fp}", f"FN: {CI_fn}", f"TN: {CI_tn}"],  # CI
    [f"TP: {CN_tp}", f"FP: {CN_fp}", f"FN: {CN_fn}", f"TN: {CN_tn}"]   # CN
])

# Plot the confusion matrix
plt.figure(figsize=(5, 4))
sns.heatmap(np.zeros_like(confusion_matrix, dtype=float), annot=confusion_matrix, fmt='', cmap='Blues',
            xticklabels=['TP', 'FP', 'FN', 'TN'], yticklabels=['AD', 'CI', 'CN'])

# Add titles and labels
plt.title('Confusion Matrix for AD, CI, and CN (with TP, FP, FN, TN)')
plt.xlabel('Metrics')
plt.ylabel('Classes')
plt.show()
```

**Classification Report Bar Chart**

```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the classification report
classes = ['Alzheimer Disease', 'Cognitive Impairment', 'Cognitively Normal']
precision = [0.9866666, 0.9758241758241758, 0.9775784753363229]
recall = [0.9866666, 0.9866666, 0.9667405764966741]
f1_score = [0.986666, 0.9812154696132597, 0.9721293199554069]


x = np.arange(len(classes))  # the label locations
width = 0.2  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, precision, width, label='Precision')
rects2 = ax.bar(x, recall, width, label='Recall')
rects3 = ax.bar(x + width, f1_score, width, label='F1-Score')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Classes')
ax.set_ylabel('Scores')
ax.set_title('Classification Report Metrics by Class')
ax.set_xticks(x)
ax.set_xticklabels(classes)
ax.legend()

fig.tight_layout()
plt.show()
```
